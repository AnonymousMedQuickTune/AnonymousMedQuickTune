```python
# read a json file and show the data
import json

import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd

def extract_values(data_list, key):
    values = []
    for data in data_list:
        for fold in data['folds']:
            values.append(fold[key])
    return np.array(values)

def load_file(file_path):
    # Open the file and load its content
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data

def load_files(file_paths):
    data_list = []
    for file_path in file_paths:
        with open(file_path, 'r') as file:
            data = json.load(file)
            data_list.append(data)
    return data_list

def load_files_config(file_paths):
    all_data = []
    for file_path in file_paths:
        with open(file_path, 'r') as f:
            data = json.load(f)
            configurations = load_configurations(data)
            all_data.append(configurations)
    return all_data

def load_configurations(file):
    configurations = []
    for configs in file:
        configurations.append(configs[0])
    return configurations

def plot_configs(data):
    # Variables
    name = "Configurations"
    value = "Set value"
    # Extracting values for each key
    bn_size = [item['bn_size'] for item in data]
    dropout_prob = [item['dropout_prob'] for item in data]
    growth_rate = [item['growth_rate'] for item in data]
    init_features = [item['init_features'] for item in data]

    # Creating subplots
    fig, axs = plt.subplots(4, 1, figsize=(8, 12))

    # Plotting bn_size
    axs[0].plot(range(1, len(bn_size) + 1), bn_size, marker='o')
    axs[0].set_title('Bottleneck size')
    axs[0].set_xlabel(name)
    axs[0].set_ylabel(value)
    axs[0].set_xticks(range(1, len(bn_size) + 1))

    # Plotting dropout_prob
    axs[1].plot(range(1, len(dropout_prob) + 1), dropout_prob, marker='o')
    axs[1].set_title('Dropout')
    axs[1].set_xlabel(name)
    axs[1].set_ylabel(value)
    axs[1].set_xticks(range(1, len(dropout_prob) + 1))

    # Plotting growth_rate
    axs[2].plot(range(1, len(growth_rate) + 1), growth_rate, marker='o')
    axs[2].set_title('Growth rate')
    axs[2].set_xlabel(name)
    axs[2].set_ylabel(value)
    axs[2].set_xticks(range(1, len(growth_rate) + 1))

    # Plotting init_features
    axs[3].plot(range(1, len(init_features) + 1), init_features, marker='o')
    axs[3].set_title('Initial features')
    axs[3].set_xlabel(name)
    axs[3].set_ylabel(value)
    axs[3].set_xticks(range(1, len(init_features) + 1))

    # Adjusting layout
    plt.tight_layout()

    # Save plots to a file
    plt.savefig('./output/config_hyperparameter.png')

    # Close the plot to avoid displaying it
    plt.close()

def plot_multiple_configs(data1, data2):
    # Define colors
    color1 = '#4666FF'  # Pastel dark blue
    color2 = '#FF7F50'  # Pastel dark orange

    # Darker colors for 'X' marks
    dark_blue_color = '#00008B'  # Even darker blue
    dark_orange_color = '#FF4500'  # Even darker orange

    # Variables
    name = "Configurations"
    value = "Set value"
    file_1 = "Lipo"
    file_2 = "Lipo" 
    # Extracting values for each key from both datasets
    bn_size1 = [item['bn_size'] for item in data1]
    bn_size2 = [item['bn_size'] for item in data2]
    dropout_prob1 = [item['dropout_prob'] for item in data1]
    dropout_prob2 = [item['dropout_prob'] for item in data2]
    growth_rate1 = [item['growth_rate'] for item in data1]
    growth_rate2 = [item['growth_rate'] for item in data2]
    init_features1 = [item['init_features'] for item in data1]
    init_features2 = [item['init_features'] for item in data2]

    # Creating subplots
    fig, axs = plt.subplots(4, 1, figsize=(8, 12))

    # Plotting bn_size
    axs[0].plot(range(1, len(bn_size1) + 1), bn_size1, marker='o', label=file_1, color=color1)
    axs[0].plot(range(1, len(bn_size2) + 1), bn_size2, marker='o', label=file_2, color=color2)
    axs[0].plot(1, bn_size1[0], marker='X', color=dark_blue_color, markersize=8)
    axs[0].plot(2, bn_size2[1], marker='X', color=dark_orange_color, markersize=8)
    axs[0].set_title('Bottleneck size')
    axs[0].set_xlabel(name)
    axs[0].set_ylabel(value)
    axs[0].set_xticks(range(1, len(bn_size1) + 1))
    axs[0].legend()

    # Plotting dropout_prob
    axs[1].plot(range(1, len(dropout_prob1) + 1), dropout_prob1, marker='o', label=file_1, color=color1)
    axs[1].plot(range(1, len(dropout_prob2) + 1), dropout_prob2, marker='o', label=file_2, color=color2)
    axs[1].plot(1, dropout_prob1[0], marker='X', color=dark_blue_color, markersize=8)
    axs[1].plot(2, dropout_prob2[1], marker='X', color=dark_orange_color, markersize=8)
    axs[1].set_title('Dropout')
    axs[1].set_xlabel(name)
    axs[1].set_ylabel(value)
    axs[1].set_xticks(range(1, len(dropout_prob1) + 1))
    axs[1].legend()

    # Plotting growth_rate
    axs[2].plot(range(1, len(growth_rate1) + 1), growth_rate1, marker='o', label=file_1, color=color1)
    axs[2].plot(range(1, len(growth_rate2) + 1), growth_rate2, marker='o', label=file_2, color=color2)
    axs[2].plot(1, growth_rate1[0], marker='X', color=dark_blue_color, markersize=8)
    axs[2].plot(2, growth_rate2[1], marker='X', color=dark_orange_color, markersize=8)
    axs[2].set_title('Growth rate')
    axs[2].set_xlabel(name)
    axs[2].set_ylabel(value)
    axs[2].set_xticks(range(1, len(growth_rate1) + 1))
    axs[2].legend()

    # Plotting init_features
    axs[3].plot(range(1, len(init_features1) + 1), init_features1, marker='o', label=file_1, color=color1)
    axs[3].plot(range(1, len(init_features2) + 1), init_features2, marker='o', label=file_2, color=color2)
    axs[3].plot(1, init_features1[0], marker='X', color=dark_blue_color, markersize=8)
    axs[3].plot(2, init_features2[1], marker='X', color=dark_orange_color, markersize=8)
    axs[3].set_title('Initial features')
    axs[3].set_xlabel(name)
    axs[3].set_ylabel(value)
    axs[3].set_xticks(range(1, len(init_features1) + 1))
    axs[3].legend()

    # Adjusting layout
    plt.tight_layout()

    # Save plots to a file
    plt.savefig('./output/config_hyperparameter.png')

    # Close the plot to avoid displaying it
    plt.close()

def plot_per_fold(data):
    # Extracting the values for each plot
    training_loss = data['folds'][0]['training_loss']
    validation_loss = data['folds'][0]['validation_loss']
    training_auc = data['folds'][0]['training_auc']
    validation_auc = data['folds'][0]['validation_auc']

    # Plotting loss values
    plt.figure(figsize=(10, 5))
    plt.plot(training_loss, label='Training Loss', marker='o')
    plt.plot(validation_loss, label='Validation Loss', marker='o')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/train_val_loss.png')
    plt.close()

    # Plotting AUC values
    plt.figure(figsize=(10, 5))
    plt.plot(training_auc, label='Training AUC', marker='o')
    plt.plot(validation_auc, label='Validation AUC', marker='o')
    plt.title('Training and Validation AUC')
    plt.xlabel('Epoch')
    plt.ylabel('AUC')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/train_val_auc.png')
    plt.close()

def plot_fold_per_config(data_list):

    # Define styles and labels
    styles = ['-', '-', '-', '-', '-']
    shades = [0.1, 0.2, 0.3, 0.4]  # Transparency levels for shaded lines
    labels = ['Config 1', 'Config 2', 'Config 3', 'Config 4', 'Config 5']
    colors = ['#4B9E9E', '#C08585', '#7B8B60', '#6A79A6', '#8B6F9A']  # Blue and Green colors

    # Plotting loss values
    plt.figure(figsize=(10, 5))
    for i, data in enumerate(data_list):
        for fold in data['folds']:
            # plt.plot(fold['training_loss'], label=f'{labels[i]} Training Loss', linestyle=styles[i % len(styles)], alpha=1.0)
            plt.plot(fold['validation_loss'], label=f'{labels[i]}', linestyle=styles[i], color=colors[i],  alpha=1.0)
    plt.title('Validation Loss - Lipo Dataset')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/train_val_loss.png')
    plt.close()

    # Plotting AUC values
    plt.figure(figsize=(10, 5))
    for i, data in enumerate(data_list):
        for fold in data['folds']:
            # plt.plot(fold['training_loss'], label=f'{labels[i]} Training Loss', linestyle=styles[i % len(styles)], alpha=1.0)
            plt.plot(fold['validation_auc'], label=f'{labels[i]}', linestyle=styles[i],color=colors[i], alpha=1.0)
    plt.title('Validation AUC - Lipo Dataset')
    plt.xlabel('Epoch')
    plt.ylabel('AUC')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/train_val_auc.png')
    plt.close()

def plot_config_per_fold(data_list):
     # Define styles and labels
    styles = ['-', '-']
    labels = ['Config 1', 'Config 2']
    colors = ['#FFA07A', '#779ECB']  # Blue and Green colors

    # Plotting loss values
    plt.figure(figsize=(10, 5))
    for i, data in enumerate(data_list):
        for fold in data['folds']:
            # plt.plot(fold['training_loss'], label=f'{labels[i]} Training Loss', linestyle=styles[i % len(styles)], alpha=1.0)
            plt.plot(fold['validation_loss'], label=f'{labels[i]}', linestyle=styles[i], color=colors[i],  alpha=1.0)
    plt.title('Validation Loss - Fold 1')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/train_val_loss.png')
    plt.close()

    # Plotting AUC values
    plt.figure(figsize=(10, 5))
    for i, data in enumerate(data_list):
        for fold in data['folds']:
            # plt.plot(fold['training_loss'], label=f'{labels[i]} Training Loss', linestyle=styles[i % len(styles)], alpha=1.0)
            plt.plot(fold['validation_auc'], label=f'{labels[i]}', linestyle=styles[i],color=colors[i], alpha=1.0)
    plt.title('Validation AUC - Fold 1')
    plt.xlabel('Epoch')
    plt.ylabel('AUC')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/train_val_auc.png')
    plt.close()

def plot_config_mean_std(data_list):
    # Extract values
    training_loss_values = extract_values(data_list, 'training_loss')
    validation_loss_values = extract_values(data_list, 'validation_loss')
    training_auc_values = extract_values(data_list, 'training_auc')
    validation_auc_values = extract_values(data_list, 'validation_auc')

    # Calculate mean and standard deviation
    training_loss_mean = np.mean(training_loss_values, axis=0)
    training_loss_std = np.std(training_loss_values, axis=0)
    validation_loss_mean = np.mean(validation_loss_values, axis=0)
    validation_loss_std = np.std(validation_loss_values, axis=0)

    training_auc_mean = np.mean(training_auc_values, axis=0)
    training_auc_std = np.std(training_auc_values, axis=0)
    validation_auc_mean = np.mean(validation_auc_values, axis=0)
    validation_auc_std = np.std(validation_auc_values, axis=0)

    # Plotting loss values
    plt.figure(figsize=(10, 5))
    epochs = range(1, len(training_loss_mean) + 1)
    plt.plot(epochs, training_loss_mean, label='Training Loss Mean', color='blue')
    plt.fill_between(epochs, training_loss_mean - training_loss_std, training_loss_mean + training_loss_std, color='blue', alpha=0.2)
    plt.plot(epochs, validation_loss_mean, label='Validation Loss Mean', color='orange')
    plt.fill_between(epochs, validation_loss_mean - validation_loss_std, validation_loss_mean + validation_loss_std, color='orange', alpha=0.2)
    plt.title('Training and Validation Loss Mean and Std')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/mean_std_loss.png')
    plt.close()

    # Plotting AUC values
    plt.figure(figsize=(10, 5))
    plt.plot(epochs, training_auc_mean, label='Training AUC Mean', color='blue')
    plt.fill_between(epochs, training_auc_mean - training_auc_std, training_auc_mean + training_auc_std, color='blue', alpha=0.2)
    plt.plot(epochs, validation_auc_mean, label='Validation AUC Mean', color='orange')
    plt.fill_between(epochs, validation_auc_mean - validation_auc_std, validation_auc_mean + validation_auc_std, color='orange', alpha=0.2)
    plt.title('Training and Validation AUC Mean and Std')
    plt.xlabel('Epoch')
    plt.ylabel('AUC')
    plt.legend()
    plt.tight_layout()
    plt.savefig('./cv_diagrams/mean_std_auc.png')
    plt.close()

def main():

    # For configurations
    # config_metrics_file = './cross_validation/configuration_fold_metrics_11.json' 
    #config_metrics_file = ['./cross_validation/configuration_fold_metrics_11.json',
    #              './cross_validation/configuration_fold_metrics_12.json' ]
    # configs_metrics = load_file(config_metrics_file)
    #configs = load_files_config(config_metrics_file)
    #plot_multiple_configs(configs[0], configs[1])

    # Per fold plotting
    # fold_metrics_file = './metrics/fold_metrics_20240627_27.json' 
    # fold_metrics = load_file(fold_metrics_file)
    # plot_per_fold(fold_metrics)

    # One configuration, multiple folds (DEleted configurations are accepted)
    file_paths = ['../paradiagramar/lipo1.json', '../paradiagramar/lipo2.json'
                  , '../paradiagramar/lipo3.json', '../paradiagramar/lipo4.json'
                  , '../paradiagramar/lipo5.json']
    multiple_folds = load_files(file_paths)
    plot_fold_per_config(multiple_folds)

    # One fold, multiple configuraitons (Here no deleted configurations)
    # file_paths = ['./metrics/fold_metrics_20240627_24.json',
    #               './metrics/fold_metrics_20240627_28.json']
    # multiple_configs = load_files(file_paths)
    # plot_config_per_fold(multiple_configs)

    # Mean and stardard deviation per configuration (Only accepted configurations)
    #file_paths = ['./metrics/fold_metrics_20240628_56.json',
    #              './metrics/fold_metrics_20240627_25.json']
    #multiple_configs = load_files(file_paths)
    #plot_config_mean_std(multiple_configs)

if __name__ == "__main__":
    main()
```
